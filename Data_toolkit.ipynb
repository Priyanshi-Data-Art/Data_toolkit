np.array:

Creates a new NumPy array from any object (lists, tuples, etc.).

Copies the data by default, ensuring that the new array is independent of the original.


```python
import numpy as np
arr1 = np.array([1, 2, 3])  # Creates a new array
```

np.asarray:

Converts input to an array, but does not copy the data if the input is already an array of the same type.

More efficient when you want to convert lists or tuples to an array without unnecessary copying.


```python
arr2 = np.asarray(arr1)  # No copy is made if arr1 is already a NumPy array
```

np.asanyarray:

Similar to np.asarray, but allows the input to be converted into any subclass of ndarray, including masked arrays and other array types.

Useful when you want to ensure the output is at least an array but are okay with subclasses.


```python
arr3 = np.asanyarray(arr1)  # Will return a subclass if the input is a subclass
```

2. Differences between Deep Copy and Shallow Copy
**Shallow Copy:**

Creates a new object, but inserts references into it to the objects found in the original.

Changes made to mutable objects in the shallow copy will reflect in the original object.

Example using copy module:

```python
import copy
original = [[1, 2, 3], [4, 5, 6]]
shallow = copy.copy(original)
shallow[0][0] = 10
print(original)  # Output: [[10, 2, 3], [4, 5, 6]] (original is affected)
```

**Deep Copy:**

Creates a new object and recursively adds copies of nested objects found in the original.

Changes made to mutable objects in the deep copy do not affect the original object.

Example using copy module:

```python
import copy
original = [[1, 2, 3], [4, 5, 6]]
deep = copy.deepcopy(original)
deep[0][0] = 10
print(original)  # Output: [[1, 2, 3], [4, 5, 6]] (original remains unchanged)
```

14). Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a date range
date_range = pd.date_range(start='2023-01-01', periods=30, freq='D')

# Step 2: Create random temperature and humidity data
np.random.seed(42)  # For reproducibility
temperature = np.random.uniform(low=15, high=30, size=len(date_range))  # Random temperatures
humidity = np.random.uniform(low=40, high=100, size=len(date_range))  # Random humidity

# Step 3: Create a DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Step 4: Plot the data
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot Temperature on the left y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (Â°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')
plt.grid()

# Show the plot
plt.show()
```

15). Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Step 1: Create a NumPy array with 1000 samples from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)  # Normal distribution with mean=0, std=1

# Step 2: Plot a histogram of the data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Histogram')

# Step 3: Overlay the PDF of the normal distribution
xmin, xmax = plt.xlim()  # Get the x limits of the histogram
x = np.linspace(xmin, xmax, 100)  # Generate x values for PDF
p = norm.pdf(x, loc=0, scale=1)  # Calculate the PDF for the normal distribution
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')

# Step 4: Label the axes and title
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency/Probability', fontsize=14)
plt.title('Histogram with PDF Overlay', fontsize=16)
plt.legend()  # Show the legend
plt.grid()

# Step 5: Show the plot
plt.show()
```

17). Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create two random arrays
np.random.seed(42)  # For reproducibility
x = np.random.randn(100)  # 100 random values from a normal distribution
y = np.random.randn(100)  # 100 random values from a normal distribution

# Step 2: Create a DataFrame to store the data
data = pd.DataFrame({'X': x, 'Y': y})

# Step 3: Determine the quadrant for each point
data['Quadrant'] = np.where((data['X'] > 0) & (data['Y'] > 0), 'Quadrant I',
                            np.where((data['X'] < 0) & (data['Y'] > 0), 'Quadrant II',
                                     np.where((data['X'] < 0) & (data['Y'] < 0), 'Quadrant III',
                                              'Quadrant IV')))

# Step 4: Create the scatter plot using Seaborn
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='Set1', s=100)

# Step 5: Customize the plot
plt.axhline(0, color='gray', lw=0.8, ls='--')  # Add horizontal line at y=0
plt.axvline(0, color='gray', lw=0.8, ls='--')  # Add vertical line at x=0
plt.title('Quadrant-wise Scatter Plot', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.legend(title='Quadrants')
plt.grid()

# Step 6: Show the plot
plt.show()
```

18). With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function

```python
from math import pi
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the data
x = np.linspace(0, 2 * pi, 100)  # 100 points from 0 to 2Ï€
y = np.sin(x)  # Sine of x

# Create a Bokeh figure
p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add grid lines
p.grid.grid_line_alpha = 0.5

# Plot the sine wave
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')

# Show the legend
p.legend.location = 'top_left'

# Show the plot
output_notebook()  # Use this if you're running in a Jupyter notebook
show(p)
```

19). Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(categories))

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a Bokeh figure
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values', height=400)  # Use height instead of plot_height

# Add bars
bars = p.vbar(x='categories', top='values', source=source, width=0.9,
              fill_color='blue', line_color='black')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Show the plot
output_notebook()  # Use this if you're running in a Jupyter notebook
show(p)
```

20). Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot

```python
import numpy as np
import plotly.graph_objects as go

# Generate random data
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.random.rand(100)  # 100 random y-values

# Create a line plot
fig = go.Figure()

# Add a trace for the line
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Label the axes and set the title
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show the plot
fig.show()
```
